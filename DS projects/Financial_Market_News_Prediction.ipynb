# **Financial Market News Prediction using Random Forest Classifier**

## **Objective**
The goal of this project is to predict financial market movements (up/down) based on daily news headlines using a machine learning model. We aim to classify whether the market will go up or down based on news data using a **Random Forest Classifier**.

## **Data Source**
- The dataset is sourced from a CSV file available at: [Financial Market News Dataset](https://github.com/YBIFoundation/Dataset/blob/main/Financial%20Market%20News.csv)

## **Import Libraries**

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

## **Data Import**

```python
df = pd.read_csv('https://raw.githubusercontent.com/YBIFoundation/Dataset/main/Financial%20Market%20News.csv', encoding="ISO-8859-1")
```

## **Data Preprocessing**

```python
# Combining all news columns into one for each row
news = []
for row in range(0, len(df.index)):
    news.append(' '.join(str(x) for x in df.iloc[row, 2:27]))
```

```python
Cv = CountVectorizer(lowercase=True, ngram_range=(1,1))
x = Cv.fit_transform(news)
```

```python
x.shape
```

## **Data Target Variables**

```python
y = df['Label']
y.shape
```

## **Train-Test Split**

```python
# Splitting data into training and testing sets (70% train, 30% test)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, stratify=y, random_state=2529)
```

## **Modeling**

```python
# Model: Random Forest Classifier
rf = RandomForestClassifier(n_estimators=200)
rf.fit(x_train, y_train)
```

## **Model Evaluation**

```python
# Making predictions on the test set
y_pred = rf.predict(x_test)
```

```python
# Evaluation using confusion matrix and classification report
confusion_matrix(y_test, y_pred)
print(classification_report(y_test, y_pred))
```

## **Explanation**
1. **Text Preprocessing**: All the news columns were merged into a single text feature for each day. This is done to allow the text to be vectorized into a format that can be used by the machine learning model.

2. **Text Vectorization**: We used **CountVectorizer**, which implements a **Bag of Words (BoW)** model to convert the text into numerical vectors. This model transforms each document (in this case, a day’s news) into a vector of word counts.

3. **Model**: A **Random Forest Classifier** with 200 estimators was trained on the training data. Random Forest is an ensemble learning method that builds multiple decision trees and combines their output to improve accuracy and prevent overfitting.

4. **Model Performance**: The model is evaluated using precision, recall, F1-score, and accuracy metrics. The classification report provides a detailed breakdown of the performance for each class (0 and 1).

5. **Application**: This approach can be applied to real-time financial news to predict market movements, aiding traders and investors in decision-making processes.
